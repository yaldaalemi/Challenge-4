```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_count = school_data_complete["school_name"].nunique()
student_count = school_data_complete["student_name"].count()
budget = school_data_complete["budget"].sum()
math = school_data_complete["math_score"].mean()
reading = school_data_complete["reading_score"].mean()
math_pass = school_data_complete.loc[:,["math_score"]] >= 70
math_pass = math_pass.loc[math_pass["math_score"]==True,["math_score"]]
math_percentage = math_pass["math_score"].count()/school_data_complete["math_score"].count()*100
reading_pass = school_data_complete.loc[:,["reading_score"]] >= 70
reading_pass = reading_pass.loc[reading_pass["reading_score"]==True,["reading_score"]]
reading_percentage = reading_pass["reading_score"].count()/school_data_complete["reading_score"].count()*100
both_pass = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70),\
                                     "student_name"]
both_percentage = both_pass.count()/school_data_complete["student_name"].count()*100
dicts = [{"school count":school_count,"student count":student_count, "total budget":budget, "average math score": math,\
          "average reading score":reading, "%Passed math":math_percentage, "%Passed reading":reading_percentage,\
          "%passed both":both_percentage}]
district_df = pd.DataFrame(dicts)
district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
school = school_data_complete.groupby(["school_name"])
student = school.count().head(15)
sum = school.sum().head(15)
type = school.first().head(15)
mean = school.mean().head(15)
Final = pd.DataFrame()
Final["School Type"] = type["type"]
Final["Total Students"] = student["student_name"]
Final["Total School Budget"] = sum["budget"]/Final["Total Students"]
Final["Per Student Budget"] = Final["Total School Budget"]/Final["Total Students"]
math_pass = school_data_complete.loc[school_data_complete["math_score"]>=70,:]
math_pass_gp = math_pass.groupby(["school_name"]).count()
math_percentage = math_pass_gp["math_score"]/Final["Total Students"]*100
reading_pass = school_data_complete.loc[school_data_complete["reading_score"]>=70,:]
reading_pass_gp = reading_pass.groupby(["school_name"]).count()
reading_percentage = reading_pass_gp["reading_score"]/Final["Total Students"]*100
both_pass = school_data_complete.loc[(school_data_complete["reading_score"]>=70)&(school_data_complete["math_score"]>=70),:]
both_pass_gp = both_pass.groupby(["school_name"]).count()
both_percentage = both_pass_gp["reading_score"]/Final["Total Students"]*100
Final["Average Math Score"] = mean["math_score"]
Final["Average Reading Score"] = mean["reading_score"]
Final["% Passing Math"] = math_percentage
Final["% Passing reading"] = reading_percentage
Final["% Overall Passing"] = both_percentage
Final.head(20)
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
sort = Final.sort_values(["% Overall Passing"], ascending=False)
sort.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
sort_bad = Final.sort_values(["% Overall Passing"])
sort_bad.head(5)
```

## Math Scores by Grade

  * Create a pandas series for each grade. Hint: use a conditional statement.

  * Combine the series into a dataframe

```python
ninth =  school_data_complete.loc[school_data_complete["grade"]=="9th",["school_name","math_score"]]
ninth_gp = ninth.groupby(["school_name"]).mean()
tenth =  school_data_complete.loc[school_data_complete["grade"]=="10th",["school_name","math_score"]]
tenth_gp = tenth.groupby(["school_name"]).mean()
eleventh =  school_data_complete.loc[school_data_complete["grade"]=="11th",["school_name","math_score"]]
eleventh_gp = eleventh.groupby(["school_name"]).mean()
twelfth =  school_data_complete.loc[school_data_complete["grade"]=="12th",["school_name","math_score"]]
twelfth_gp = twelfth.groupby(["school_name"]).mean()
Math = pd.DataFrame()
Math["9th"]= ninth_gp
Math["10th"]= tenth_gp
Math["11th"]= eleventh_gp
Math["12th"]= twelfth_gp
Math.head(10)
```
